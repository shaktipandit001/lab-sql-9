```python
def compare(rental_dates):
    month_counts = {'May': 0, 'June': 0}

    for date in rental_dates:
        if date.month == 6:
            month_counts['June'] += 1
        else:
            month_counts['May'] += 1

    if month_counts['June'] > month_counts['May']:
        print('more films are borrowed in the month of June.')
    elif month_counts['June'] < month_counts['May']:
        print('more films are borrowed in the month of May.')
    else:
        print('equal amounts of films are borrowed in both months.')
```
